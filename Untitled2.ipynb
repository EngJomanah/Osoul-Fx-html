```python
import ccxt
import pandas as pd

# إعداد Binance Testnet
exchange = ccxt.binance({
    'apiKey': 'ضع_هنا_API_KEY',
    'secret': 'ضع_هنا_API_SECRET',
    'enableRateLimit': True,
})
exchange.set_sandbox_mode(True)  # يفعّل وضع Testnet

# جلب آخر 500 شمعة لزوج BTC/USDT على فاصل 5 دقائق
klines = exchange.fetch_ohlcv('BTC/USDT', timeframe='5m', limit=500)

# تحويلها إلى DataFrame
df = pd.DataFrame(klines, columns=['timestamp','open','high','low','close','volume'])
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
for col in ['open','high','low','close','volume']:
    df[col] = df[col].astype(float)

df.tail()
```

```python
import ta

df['sma20'] = ta.trend.sma_indicator(df['close'], window=20)
df['rsi14'] = ta.momentum.rsi(df['close'], window=14)
df['atr14'] = ta.volatility.average_true_range(df['high'], df['low'], df['close'], window=14)
df = df.dropna()
df.tail()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

# صنع تسمية بسيطة: شراء إذا سعر الشمعة بعد 3 شموع أعلى من 0.2%
df['future_close'] = df['close'].shift(-3)
df['future_return'] = (df['future_close'] - df['close']) / df['close']
df['target'] = (df['future_return'] > 0.002).astype(int)

features = ['sma20','rsi14','atr14']
X = df[features].iloc[:-3]
y = df['target'].iloc[:-3]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
model = XGBClassifier(n_estimators=100, max_depth=3, use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("دقة النموذج:", accuracy_score(y_test, y_pred))

# حفظ النموذج
joblib.dump(model, "model.joblib")
```
